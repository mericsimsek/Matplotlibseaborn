```python
import seaborn as sns
import pandas as pd
```

```python
dia=sns.load_dataset("diamonds")
df=dia.copy()
```

```python
df["cut"].value_counts().plot.barh().set_title("Cut değişken frekansı")
```

```python
sns.barplot(x="cut",y = df.cut.index , data=df)
```

```python
sns.catplot(x="cut",y="price",data=df)
```

```python
sns.barplot(x="cut",y="price",hue="color",data=df)
```

```python
df.groupby(["cut","color"])["price"].mean()#cut ve colora göre kır fiyatların ort. al
```

```python
sns.distplot(df.price,kde=False)
```

```python
sns.distplot(df.price,bins=10,kde=False)
```

```python
sns.distplot(df.price)
```

```python
sns.distplot(df.price,hist=False) #histogram kapatılır yoğunuk açık
```

```python
df["price"].describe()
```

```python
sns.kdeplot(df.price,shade=True) #histogram kapatılır yoğunuk açık
```

```python
sns.FacetGrid(df,hue="cut",height=6,xlim=(1,10000)).map(sns.kdeplot ,"price", shade =True).add_legend()
##yukarıdaki gibi yine yoğunluk grafiği oluşturacaz  add legend bilgi ekler
#sns kdeplot sayesinde ama bunun üzerine boyut eklemek yani çaprazlamak istiyoruz (map ile eşlememiz gerekir)
```

```python
sns.catplot(x="cut",y="price",hue="color",kind="point" , data=df)   #point noktalar ile ifade hue renk tonu demek
```
