```python
import seaborn as sns
fm=sns.load_dataset("fmri")
df=fm.copy()
df.describe()
```

```python
df.groupby("signal")["timepoint"].describe()
```

```python
sns.lineplot(x="timepoint",y="signal",data=df)
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",data=df)
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",style="event",data=df)
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",style="event",markers=True , dashes=True,data=df)
```
